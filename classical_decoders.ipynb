# Decoding messages

When a codeword is sent through a physical channel, some bits may be corrupted. As a result, the received message might differ from the original codeword. Without knowing exactly where the errors occured, we cannot be certain what the original codeword was. So instead we try to find the codeword that is most likely to have been sent. To do this we use a probablistic model of the transmission channel.

### Discrete channels

**Definition.** [Discrete channel] Let $\mathcal{X}$ and $\mathcal{Y}$ be finite sets and let $X$ and $Y$ be random variables that take values in $\mathcal{X}$ and $\mathcal{Y}$ respectively. A _discrete channel_ is a set of probability mass functions 
$$p_{Y \mid X} = \{ p_{Y \mid X=x} \colon \mathcal{Y} \to [0,1] \mid x \in \mathcal{X}\}.$$
The discrete channel is denoted $(\mathcal{X}, p_{Y \mid X}, \mathcal{Y})$.

For each input $x \in \mathcal{X}$ and output $y \in \mathcal{Y}$, we interpret the value 
$$p_{Y \mid X=x}(y) = p(Y=y \mid X=x)$$
as the probability of observing output $y$ given that input $x$ was sent through the channel.


**Definition.** [Binary channel] A _binary_ channel is a discrete channel with $\mathcal{X} = \{0,1\} = \mathcal{Y}$.

In practice, we use the same physical channel multiple times. Ideally we would like to treat each use of the channel with the same probability distribution, but a priori it is unclear whether outputs depend on past or future inputs and outputs. For each $n \in \mathbb{N}$, define random variables $X^n$ and $Y^n$ that take values in $\mathcal{X}^n$ and $\mathcal{Y}^n$ respectively. For sequences
$$ x^n = (x_1,x_2,\dots,x_n) \in \mathcal{X} \;\; \text{ and } \;\; y^n = (y_1,y_2,\dots,y_n) \in \mathcal{Y}, $$
the chain rule gives the factorisation
$$ p(Y^n = y^n \mid X^n = x^n) = \prod_{i=1}^n p(Y_i = y_i \mid X^n = x^n, Y^{i-1} = (y_1,y_2,\dots,y_{i-1})). $$
Thus, the value of $Y_i$ may in general depend on all inputs $X^n$, all previous outputs $Y^{i-1}$, and the time index $i$. We introduce assumptions to simplify the dependence

---

#### Assumptions

**Causal.**

A channel $(\mathcal{X}, p_{Y \mid X} \mathcal{Y})$ is _causal_ if the current output does not depend on future inputs. That is, for each $i, n \in \mathbb{N}$ with $i \leq n$ and sequences
$$ x^n = (x_1,x_2,\dots,x_n) \in \mathcal{X} \;\; \text{ and } \;\; y^n = (y_1,y_2,\dots,y_n) \in \mathcal{Y}, $$
 we have that 
$$ p(Y_i = y_i \mid X^n = x^n, Y^{i-1} = y^{i-1}) = p(Y_i = y_i \mid X^i = x^i, Y^{i-1} = y^{i-1}). $$

**Memoryless.**

A channel $(\mathcal{X}, p_{Y \mid X} \mathcal{Y})$ is _memoryless_ if the current output does not depend on previous inputs or outputs. That is, for each $i, n \in \mathbb{N}$ with $i \leq n$ and sequences
$$ x^n = (x_1,x_2,\dots,x_n) \in \mathcal{X} \;\; \text{ and } \;\; y^n = (y_1,y_2,\dots,y_n) \in \mathcal{Y}, $$
 we have that 
$$ p(Y_i = y_i \mid X^i = x^i, Y^{i-1} = y^{i-1}) = p(Y_i = y_i \mid X_i = x_i). $$

**Stationary.** 

A channel $(\mathcal{X}, p_{Y \mid X} \mathcal{Y})$ is _stationary_ if the conditional distribution of $Y_i$ on $X_i$ does not depend on $i$. That is, for all integers $i,j,n \in \mathbb{N}$ with $i,j \leq n$, inputs $x \in \mathcal{X}$ and outputs $y \in \mathcal{Y}$, we have that 
$$ p(Y_i = y \mid X_i = x) =  p(Y_j = y \mid X_j = x).$$

---

Therefore, for a discrete channel that is causal, memoryless and stationary the conditional distribution of $Y^n$ on $X^n$ is fully determined by the conditional distribution of $Y$ on $X$. In particular, recall by the chain rule we have
$$p(Y^n = y^n \mid X^n = x^n) = \prod_{i=1}^n p(Y_i = y_i \mid X^n = x^n, Y^{i-1} = y^{i-1}).$$
Then causality gives
$$p(Y_i = y_i \mid X^n = x^n, Y^{i-1} = y^{i-1}) = p(Y_i = y_i \mid X^i = x^i, Y^{i-1} = y^{i-1})).$$
Memorylessness gives
$$p(Y_i = y_i \mid X^i = x^i, Y^{i-1} = y^{i-1})) = p(Y_i = y_i \mid X_i = x_i).$$
Finally, stationary gives
$$p(Y_i = y_i \mid X_i = x_i) = p(Y = y_i \mid X = x_i).$$
Combining these expressions we obtain
$$p(Y^n = y^n \mid X^n = x^n) = \prod_{i=1}^n p(Y = y_i \mid X = x_i).$$


#### Binary symmetric channels

Since we wish to transmit binary codewords we will focus on binary channels. From now on all channels will be discrete, binary, causal, memoryless and stationary. 

Fix a channel $(\{0,1\}, p_{Y \mid X}, \{0,1\})$. The probability mass functions $p_{Y \mid X=0}$ and $p_{Y \mid X=1}$ are uniquely defined by the four probabilites
$$ p(Y=0 \mid X=0), \;\;\; p(Y=1 \mid X=0), \;\;\; p(Y=0 \mid X=1), \;\;\; p(Y=1 \mid X=1). $$
We can group these probabilities according to whether $X=Y$ or $X \neq Y$. A binary channel is symmetric if there exists a value $p \in [0,1]$ such that
$$p(Y = y | X = x) = \begin{cases} 1-p &\text{ if } y=x \\ p &\text{ if } y \neq x. \end{cases}$$
Therefore, for sequences $x^n = (x_i)$ and $y^n = (y_i)$ in $\{0,1\}^n$, the probability $p(Y^n = y^n | X^n = x^n)$ depends only on the number of coordinates where $x^n$ and $y^n$ differ. This is the Hamming distance between $x^n$ and $y^n$. Define
$$d = \sum_{i=1}^n \mathbb{1}_{x_i \neq y_i}.$$
Then $x^n$ and $y^n$ differ in exactly $d$ coordinates and we have
$$p(Y^n = y^n | X^n = x^n) = \prod_{i=1}^n p(Y=y_i \mid X = x_i) = p^d(1-p)^{n-d}.$$



## Maximum likelihood decoding

Once we observe an output of a binary symmetric channel, we need to ascertain which codeword was sent. When a codeword is sent through a physical channel, some bits may be corrupted. As a result, the received output might differ from the original codeword. Decoding aims to recover the original codeword despite these errors. Without knowing the exact errors that occurred, we cannot be certain that decoding will find the intended codeword. So instead we try to find the codeword that is most likely to have been sent. 

Let $\mathcal{C}$ be a code of length $n \in \mathbb{Z}_{>0}$ and suppose the we observe an output $v \in \{0,1\}^n$. Then we wish to find a codeword $\hat{c} \in \mathcal{C}$ such that 
\begin{equation*}
    p(\text{ Input} = \hat{c} \mid \text{ Output} = v) = \mathrm{max}_{c \in \mathcal{C}} \{p(\text{ Input} = c \mid \text{ Output} = v)\}.
\end{equation*}

For each bit position $i \in [n]$, define two random variables:
1. Let $C_i$ be the random variable whose value is the $i$th bit of the sent codeword. Let $C$ be the random variable $C = (C_1,C_2,\dots,C_n)$, which represents the full codeword.
2. Let $V_i$ be the random variable whose value is the $i$th bit of the observed output. Let $V$ be the random variable $V = (V_1,V_2,\dots,V_n)$ which represents the full observed output.

If we observe the output $v$, we wish to find the codeword $\hat{c} \in \mathcal{C}$ that maximises the posterior probability
\begin{equation*}
    p(C = \hat{c} \mid V = v) = \mathrm{max}_{c \in \mathcal{C}} \{p(C = c \mid V = v)\}.
\end{equation*}
By abuse of notation we will often drop the notation for the random variables $C$ and $V$ and instead of writing $p(C=c \mid V=v)$, for example, write $p(c \mid v)$.


The application of Bayes Theorem to $p(c \mid v)$ results in
$$p(c \mid v) = \frac{p(v \mid c)\cdot p(c)}{p(v)}.$$
Notice that $p(v)$ does not depend on the specific codeword $c$, and so, to maximise $p(c \mid v)$ over $c \in \mathcal{C}$ it suffices to maximise $$p(v \mid c)\cdot p(c)$$ over $\mathcal{C}$. Moreover, we make the assumption that each codeword is equally likely to be sent. That is, for all codewords $c \in \mathcal{C}$ we have that $$p(c) = \frac{1}{|\mathcal{C}|}.$$ Thus, maximising $p(v \mid c)\cdot p(c)$ over $c \in \mathcal{C}$ is equivalent to maximising $p(v \mid c)$ over $c \in \mathcal{C}$.

---
**Example.** Let us consider the code $\mathcal{C} = \{000, 111\}$. Suppose that we use a binary symmetric channel and each bit has a probability $p=\frac{1}{4}$ of flipping regardless of position. Suppose that we receive the message $m=100$ and we wish to compute the most likely codeword it came from. Since maximising $p(c \mid v)$ over $\mathcal{C}$ is equivalent to maximising $p(v \mid c)$ over $\mathcal{C}$, it suffices to compute $p(v \mid c)$ for each codeword $c \in \mathcal{C}$. That is, for $c=000$ and $c=111$.

First suppose the codeword $c$ was $000$. Then receiving the message $m=100$ means the first bit flipped and the other two did not flip (TODO: can a bit flip more than once?). In particular, we have
\begin{align*}
    p(m=100 \mid c=000) &= p(m_1=1 \mid c_1=0) \cdot p(m_2=0 \mid c_2=0) \cdot p(m_3=0 \mid c_3=0) \\ &= p(c_1 \text{ did flip}) \cdot p(c_2 \text{ did not flip}) \cdot p(c_3 \text{ did not flip}) \\ &= \frac{1}{4} \cdot \frac{3}{4} \cdot \frac{3}{4} = \frac{9}{64}. 
\end{align*}
Now suppose the codeword $c$ was $111$, then receiving the message $m=100$ means the first bit did not flip and the other two did flip, so we have
\begin{equation*}
    p(m=100 \mid c=111) = \frac{3}{4} \cdot \frac{1}{4} \cdot \frac{1}{4} = \frac{3}{64}. 
\end{equation*}
Therefore, as $p(m=100 \mid c=000)$ is larger than $p(m=100 \mid c=111)$ it is most likely that the message $m=100$ came from the codeword $c=000$.
We can compute the most likely codeword for each of the possible messages $m \in (\mathbb{Z}/2\mathbb{Z})^3$.

#### Hamming distance

We can simplify the Maximum Likelihood Decoder by using the Hamming distance. Recall maximising $\mathbb{P}(c \mid m)$ is equivalent to maximising $\mathbb{P}(m \mid c) = \prod_{i=1}^n \mathbb{P}(m_i \mid c_i)$. Let us assume we have a binary symmetric channel such that the probability the probability a bit flips is $p$. Therefore, we have that
$$
\mathbb{P}(m_i \mid c_i) =
\begin{cases}
p & \text{if } m_i \neq c_i, \\
1-p & \text{if } m_i = c_i.
\end{cases}
$$
Notice that the probabilities depend only whether the bits are the same or different. In particular, the power of $p$ in the product $\prod_{i=1}^n \mathbb{P}(m_i \mid c_i)$ is precisely the number of bits in $m$ that differ from the bits in $c$, that is the Hamming distance $d(m,c)$. Therefore, we have that 
$$\mathbb{P}(m \mid c) = \prod_{i=1}^n \mathbb{P}(m_i \mid c_i) = p^{d(m,c)}(1-p)^{n-d(m,c)}.$$
Rearranging this expression we obtain
$$\mathbb{P}(m \mid c) = \left( \frac{p}{1-p}\right)^{d(m,c)}  \cdot (1-p)^{n}.$$
Note that $(1-p)^{n}$ depends neither on the codeword $c$ nor the message $m$. Thus, maximising $\mathbb{P}(m \mid c)$ over $\mathcal{C}$ is equivalent to maximising
$$\left( \frac{p}{1-p}\right)^{d(m,c)}$$
over $\mathcal{C}$.

By convention we assume that $p < \frac{1}{2}$. If it is not, then we can assume the received message is more likely to be wrong and flip all the bits before working with them .... (Fix this phrasing!). Consequently, the value $\frac{p}{1-p}$ is less than $1$ and so to maximise $$\left( \frac{p}{1-p}\right)^{d(m,c)}$$ we must minimise $d(m,c)$. In particular, we have that the codeword $\hat{c}$ that maximises $\mathbb{P}(c \mid m)$ is precisely the codeword that minimises $d(m,c)$.

$$ d(m=100,c=000) = 1 \; \text{ and } \; d(m=100,c=111) = 2 $$
and so $c=000$ minimises the Hamming distance.

The Hamming distance makes it easier to find the most likely codeword. However, for a code of rank $k$, it still requires computing and comparing $2^k$ different numbers. Therefore, as the code gets bigger the Maximum Likelihood Decoder becomes computationally impractical. We need more efficient algorithms that estimate the most likely codeword without having to compute the Hamming distances explicitly.

```python
from linearcodes import Codeword, LinearCode, HammingCode
import random
import numpy as np
import time

def send_codeword(
        code_length: int,
        channel_probabilities: list[float],
        codeword: Codeword):
    if len(channel_probabilities) != code_length:
        raise ValueError(f"List of probabilities must be of length code_length={code_length}. Got {len(channel_probabilities)}.")
    if len(codeword) != code_length:
        raise ValueError(f"Codeword must of length code_length={code_length}. Got {len(codeword)}.")
    message = []
    for bit_idx in range(code_length):
        p = channel_probabilities[bit_idx]
        c = codeword.bits[bit_idx]
        r = np.random.uniform(0,1)
        if r < p:
            flip_bit = (c+1)%2
            message.append(flip_bit)
        else:
            message.append(c)
    return message

def hamming_distance(x: list[int], y: list[int]) -> int:
    if len(x) != len(y):
        raise ValueError("Both x and y must be the same length.")
    return sum([1 for bit_idx in range(len(x)) if x[bit_idx] != y[bit_idx]])


def minimise_hamming_distance(code_length: int,
                              codewords: list[Codeword],
                              message: list[int]) -> Codeword:
    """
    Returns the codeword that was most likely to be sent given message is received using Hamming distance.
    Args:
        - channel_probabilities (list[float]):
            entry with index i is the probability bit i will flip.
        - code (lc.LinearCode):
            code used for encoding.
        - message (list[int]):
            received message.
    """
    if len(message) != code_length:
        raise ValueError(f"Message must be of length {code_length}. Got {len(message)}.")
    most_likely_codeword = Codeword([0]*code_length)
    min_hamming_dist = code_length
    for c in codewords:       
        hamming_dist = hamming_distance(c.bits, message)
        if hamming_dist < min_hamming_dist:
            most_likely_codeword = c
            min_hamming_dist = hamming_dist
    return most_likely_codeword

code = HammingCode(3)
code_length = code.length
codewords = code.codewords
channel_probabilities = [float(0.1)]*code_length
results: dict[int, tuple[bool, float]] = {}

num_trials = 10
for trial in range(num_trials):
    print(f"\n=== Minimise Hamming distance decoder {trial+1} ===")
    codeword = random.choice(codewords)

    message = send_codeword(code_length, channel_probabilities, codeword)
    start_time = time.time()
    most_likely_codeword = minimise_hamming_distance(code_length, codewords, message)
    decode_time = time.time() - start_time

    print(f"\nRows are: sent codeword c, received message m, most likely sent codeword c'.")
    data = np.array([codeword.bits, message, most_likely_codeword.bits])
    print(data)

    decode_correct = codeword == most_likely_codeword
    print(f"\nDecoding correct? {decode_correct}")
    print(f"Actual number of errors: {sum(1 for bit_idx in range(code_length) if codeword.bits[bit_idx] != message[bit_idx])}")
    print(f"Guessed number of errors: {sum(1 for bit_idx in range(code_length) if most_likely_codeword.bits[bit_idx] != message[bit_idx])}")

    results[trial] = (decode_correct, decode_time)
    print(f"\nDecode correct? {decode_correct}")
    print(f"Time to decode = {decode_time*(10**5):.3f}x10^5")
```

## Message Passing

Maximum likelihood decoding identifies the most likely codeword for a received message by comparing it against every possible codeword. While this is theoretically optimal, it becomes infeasible for practical codes because the number of codewords grows exponentially with the code rank. Message passing is a scalable alternative. It is a graph-based algorithm where each node has local information and can communicate only with its immediate neighbours. Nodes iteratively update their knowledge based on messages received, and through repeated exchanges, information spreads across the network. Over time, global properties can be inferred even though no node initially has complete knowledge.

Message passing is a analogy of social networks. People who spend a lot of time together tend to share opinions and information. Each individual updates their understanding based on what they hear from friends. Then when they passes this updated information along to other friends. As this process iterates, ideas and knowledge propagate through the network, allowing people to build a collective understanding even without direct access to everyone’s thoughts.

We can use this type of algorithm on the Tanner graph of an error correcting code. Bit nodes and parity check nodes will exchange messages along edges of the graph, updating their beliefs about errors iteratively. First we do a concrete example of message passing.

### Think of a title

Suppose we wish to count the number of vertices in a tree. Each node can only see itself and its neighbours, so it only has local information. However, we can use message passing to combine this information and create a global picture.

Let $V$ be the set of vertices and $E$ be the set of edges of a graph $G=(V, E)$. For a vertex $u \in V$ and let $\mathcal{N}(u)$ denote the neighbourhood of the vertex $u$ in $G$. For each vertex $u \in V$, we define a sequence integers $$N_0(u), \, N_1(u), \, N_2(u), \, N_3(u) \, \dots$$ by the following algorithm:

**Initialisation:** For each vertex $u \in V$, define $N_0(u) = 1$. For each pair of neighbouring vertices $(u,v)$ define $m_1(u,v) = 1$.

**Iterative step for $i>0$:**
1. **Update vertex values:** Update the information for each vertex using the following formula
$$N_i(u) = N_{i-1}(u) + \sum_{v \in \mathcal{N}(u)} m_{i}(v,u).$$
2. **Update messages:** For each pair of neighbouring vertices $(u,v)$ define a message $m_{i+1}(u,v)$ by
    $$ m_{i+1}(u,v) = \sum_{\substack{w \in \mathcal{N}(u) \\ w \neq v}} m_{i}(w,u). $$

    The message $m_{i+1}(u,v)$ will be the information "sent" from vertex $u$ to vertex $v$. This is why we do not include $v$ in the summation index for $m_i(u,v)$. If we did we would be sending information back to $v$ that it already knows and sent to us on the previous iteration.
3. **Check stop condition:** If all the messages are $0$, stop the algorithm.
 
**Claim:** Let $G = (V,E)$ be a finite tree. Then the algorithm terminates after finitely many steps. Moreover, if the algorithm terminates on iteration $i>0$, then for each vertex $u \in V$, the value $N_i(u)$ is the number of vertices in $T$.

Before, proving this statement, let us consider a concrete example.


---
**Example.** Consider an $A_3$ graph, that is a straight line with three vertices
$$ u_1 \text{---} u_2 \text{---} u_3. $$
Each of $N_0(u_i)$ is initialised at $1$ and each message is initialised at $1$. Now let us apply the algorithm.

**Iteration 1**
1. **Update vertex values:** We have that
    $$ N_1(u_1) = N_0(u_1) + \sum_{v \in \mathcal{N}(u_1)} m_1(v,u_1) = N_0(u_1) + m_1(u_2, u_1) = 1 + 1 = 2.$$
    Similarly, we compute that $N_1(u_2) = 3$ and $N_1(u_3)=1$.
2. **Update messages:** Since $u_1$ and $u_3$ have exactly one neighbour, namely $u_2$, we have that
    $$m_2(u_1,u_2) = \sum_{\substack{w \in \mathcal{N}(u_1) \\ w \neq u_2}} m_{1}(w,u_1) = 0$$
    and similarly $m_2(u_3,u_2) = 0$. However, we have two non-zero messages given by
    $$m_2(u_2,u_1) = \sum_{\substack{w \in \mathcal{N}(u_2) \\ w \neq u_1}} m_{1}(w,u_2) = m_{1}(u_3,u_2)=1$$
    and $m_2(u_2,u_3) = m_1(u_1,u_2) = 1$.
3. **Check stop condition:** Two of the messages are non-zero.

**Iteration 2**
1. **Update vertex values:** Since $u_2$ only receives zero messages, it follows that $N_2(u_2) = N_1(u_2) = 3$. For $u_1$ we have that
    $$ N_2(u_1) = N_1(u_1) + \sum_{v \in \mathcal{N}(u_1)} m_2(v,u_1) = N_1(u_1) + m_2(u_2, u_1) = 2 + 1 = 3.$$
    Similarly, we compute that $N_2(u_3) = 3$.
2. **Define messages:** All the messages will be zero in this iteration. Again since $u_1$ and $u_3$ have exactly one neighbour, namely $u_2$, we have that $m_3(u_1,u_2) = 0$ and $m_3(u_3,u_2)=0$. Moreover, we have that
    $$m_3(u_2,u_1) = \sum_{\substack{w \in \mathcal{N}(u_2) \\ w \neq u_1}} m_{1}(w,u_2) = m_2(u_3,u_2)=0$$
    and $m_3(u_2,u_3) = m_2(u_1,u_2) = 0$.
3. **Check stop condition:** All the messages are $0$. Stop the algorithm.

The algorithm terminated on iteration $2$, so for each vertex we obtained a sequence of $3$ values, as follows
$$ u_1: \, N_0(u_1) = 1, \, N_1(u_1) = 2, \, N_2(u_1) = 3, $$
$$ u_2: \, N_0(u_2) = 1, \, N_1(u_2) = 3, \, N_2(u_2) = 3, $$
$$ u_3: \, N_0(u_3) = 1, \, N_1(u_3) = 2, \, N_2(u_3) = 3. $$
The final value of each sequence is $3$; the number of vertices in $A_3$.

Now we prove the claim.

**Claim:** Let $G = (V,E)$ be a finite tree. Then the algorithm terminates after finitely many steps. Moreover, if the algorithm terminates on iteration $i>0$, then for each vertex $u \in V$, the value $N_i(u)$ is the number of vertices in $T$.

We do this in steps, starting with an analysis on the messages.

---
**Claim:** Let $G = (V,E)$ be a tree. Fix vertices $u, v \in V$ and an integer $i>0$. Then the message $m_i(u,v)$ is the number of vertices whose minimum distance from $u$ is $i-1$ via a path that does not go through $v$.

*Proof:* We prove this by induction on the number of iterations. By definition, $m_1(u,v) = 1$. This correctly counts the vertex $u$ itself, which is at distance $0 = i-1$ from $u$ and whose minimal path does not go through $v$. Now assume the claim holds for iteration $i$ and consider iteration $i+1$. By the message passing algorithm, we have
$$
m_{i+1}(u,v) = \sum_{w \in \mathcal{N}(u), w \neq v} m_i(w,u).
$$
By the induction hypothesis, each $m_i(w,u)$ counts all vertices whose minimal path from $w$ does not go through $u$ and has length $i-1$. Concatenating each of these paths with the edge $w \to u$ gives paths of length $i$ from $u$ that still do not pass through $v$. Moreover, since $G$ is a tree, there is a unique path between any two vertices. Therefore, summing over all neighbours $w \neq v$ of $u$ correctly counts all vertices whose minimal path from $u$ has length $i$ and does not pass through $v$.

---

**Claim:** Let $G = (V,E)$ be a tree. Fix a vertex $u \in V$ and an integer $i>0$. Then $N_i(u)$ is the number of vertices that are at most $i$ steps away from $u$.

*Proof:* We prove this by induction on the number of iterations. By definition, $N_0(u) = 1$, which correctly counts only the vertex $u$ itself, as it is the only vertex at distance $0$ from $u$. Assume the statement holds for iteration $i$ and consider iteration $i+1$. By the message passing algorithm,
$$
N_{i+1}(u) = N_i(u) + \sum_{v \in \mathcal{N}(u)} m_i(v,u).
$$
By the induction hypothesis, $N_i(u)$ counts all vertices at distance at most $i$ from $u$. Moreover, for each neighbour $v \in \mathcal{N}(u)$, the message $m_i(v,u)$ counts all vertices whose minimal path from $v$ does not go through $u$ and has length $i$. Therefore, concatenating these paths with the edge $v \to u$ gives a new path to vertices at distance $i+1$ from $u$. Since $G$ is a tree, there is a unique path between any two vertices. Therefore, summing over all neighbours $v$ correctly counts all vertices that are exactly $i+1$ steps away from $u$. Adding $N_i(u)$ then counts all vertices at distance at most $i$ from $u$.

---

**Claim:** Let $G = (V,E)$ be a finite tree. Then the algorithm terminates after finitely many steps. Moreover, if the algorithm terminates on iteration $i>0$, then for each vertex $u \in V$, the value $N_i(u)$ is the number of vertices in $T$.

*Proof:* Since $G$ is a finite tree, there exists an integer $d>0$ such that the distance between any two vertices in $G$ is at most $d$. Recall, for any vertices $u$ and $v$, the message $m_{d+1}(u,v)$ counts the number of vertices whose minimal path from $u$ is of length $d$ and does not go through $v$. If such a vertex exists, then concatenating its minimal path to $u$ with the edge $u \text{---} v$ would create a path of length $d+1$; a contradiction. Therefore, for all vertices $u$ and $v$, we have $m_{d+1}(u,v)=0$ and the algorithm terminates at iteration $d$. Moreover, because $N_d(u)$ is the number of vertices at most $d$ steps away from $u$, we have that $N_d(u) = |V|$.




We can encode message passing to work on large graphs.

```python
import networkx as nx
import random

class MessagePassing:
    """
    Implements a message passing algorithm on a NetworkX tree.
    Algorithm sends integer messages, adding them at each iteration.
    
    Args:
        - num_vertices [int]: Number of vertices in the tree.
        - initial_vertex_values [list[int]]: Optional list of initial values for each vertex.
            If not given, then defaults to 1 for all vertices.
    """
    def __init__(self,
                 graph: nx.Graph,
                 initial_vertex_values: list[int] | None = None):
        self.graph = graph
        self.num_vertices = len(graph.nodes)  

        if initial_vertex_values is None:
            initial_vertex_values = [1]*self.num_vertices
        if len(initial_vertex_values) != self.num_vertices:
            raise ValueError("Length of initial_vertex_values must match num_vertices.")
        
        self.neighbours = [list(self.graph.neighbors(u)) for u in self.graph.nodes]
        self.initial_vertex_values = initial_vertex_values
        
        self.directed_edges = [(u, v) for u in self.graph.nodes for v in self.graph.neighbors(u)]
        
    @property
    def initial_messages(self) -> dict[tuple[int,int],int]:
        """
        Creates list of initial messages.
        The message from a vertex u to a neighbour v is the inital value of u.
        """
        return {edge : self.initial_vertex_values[edge[0]] for edge in self.directed_edges}

    def run(self, max_interations: int = 100, verbose: bool = False) -> tuple[int,list[int]]:
        """
        Run message passing until convergence or max_iterations reached.
        
        Args:
            - max_iterations [int]: Maximum number of iterations of the algorithm.
            - verbose [bool]: If True, prints messages and vertex values at each iteration.
        
        Returns:
            list[int]: The final vertex values after message passing.
        """
        vertex_values = self.initial_vertex_values
        messages = self.initial_messages.copy()
        for iteration in range(max_interations):
            vertex_values = self._update_vertex_values(vertex_values, messages)
            
            new_messages = self._update_messages(messages)
            if all(value == 0 for value in new_messages.values()):
                    if verbose:
                        print(f"Converged at iteration {iteration+1}.")
                    return (iteration, vertex_values)
            messages = new_messages

            if verbose:
                print(f"\nIteration {iteration+1}")
                print(f"Messages: {messages}")
                print(f"Vertex values: {vertex_values}")

        return (iteration,vertex_values)

    def _update_messages(self,
                        messages: dict[tuple[int,int], int]) -> dict[tuple[int,int], int]:
        """
        Compute new messages based on incoming messages.
        """
        new_messages = {edge : 0 for edge in self.directed_edges}
        for u, v in self.directed_edges:
            new_message = 0
            for w in self.neighbours[u]:
                if w != v:
                    try:
                        new_message += messages[(w,u)]
                    except:
                        pass
            new_messages[(u,v)] = new_message
        return new_messages
    
    def _update_vertex_values(self,
                              vertex_values: list[int],
                              messages: dict[tuple[int,int], int]) -> list[int]:
        """
        Update vertex values based on the messages received from neighbours.
        """
        new_vertex_values = vertex_values.copy()
        for u, v in self.directed_edges:
            new_vertex_values[v] += messages[(u,v)]
        return new_vertex_values

def find_number_vertices_in_random_tree(max_num_vertices: int = 50, num_iterations: int = 100):
    num_vertices = random.choice(range(max_num_vertices))
    graph = nx.generators.random_labeled_tree(num_vertices)
    mp = MessagePassing(graph)
    final_info = mp.run(num_iterations)
    return (final_info[0], final_info[1][0])        
    
for _ in range(10):
    iteration, num_vertices = find_number_vertices_in_random_tree(1000, 1000)
    print(f"After {iteration} iterations computed the tree has {num_vertices} vertices.")
```

If the graph is not a tree, then the algorithm is not guaranteed to converge in finitely many steps.

**Example (Cycle of length 3).** Consider the complete graph on $3$ vertices $V = \{u_1, u_2, u_3\}$. Then we claim that for each pair of vertices $u_k, u_\ell \in V$ and each iteration $i>0$, we have that 
$$m_i(u_k,u_\ell)=1 \; \text{ and } \; N_i(u_k) = 2i + 1.$$

*Proof:* We shall prove the statement by induction on the number of iterations $i$. Each of the vertices is initialised at $1$ which coincides with $2(0) + 1 = 1$ and the first messages are also initialised at $1$ as required. Now suppose the statement holds for iteration $i$ and consider iteration $i+1$. Since the graph is cyclically symmetric it suffices to prove the statement for just one of the messages and one of the vertex values. Let's consider $m_{i+1}(u_1,u_2)$ and $N_{i+1}(u_1)$. The updated message $m_{i+1}(u_1,u_2)$ is given by the formula
\begin{align*}
m_{i+1}(u_1, u_2) &= \sum_{w \in \mathcal{N}(u_1) \setminus \{u_2\}} m_i(w, u_1) \\ &= m_i(u_3, u_1).
\end{align*}
Moreover, by the inductive hypothesis, we have that $m_i(u_3,u_1) = 1$, and so $m_{i+1}(u_1,u_2) = 1$ as required. Now consider the value of the vertex $u_1$. We have that 
\begin{align*}
N_{i+1}(u_1) &= N_i(u_1) + \sum_{v \in \mathcal{N}(u_1)} m_{i+1}(v, u_1) \\ &= N_i(u_1) + m_{i+1}(u_2, u_1) + m_{i+1}(u_3, u_1).
\end{align*}
By the inductive hypothesis, we have that $N_i(u) = 2i+1$ and $m_i(u_2, u_1) = m_i(u_3, u_1) = 1$. Thus, we obtain
\begin{align*}
N_{i+1}(u_1) &= N_i(u_1) + m_i(u_2, u_1) + m_i(u_3, u_1) \\ &= (2i + 1) + 1 + 1 = 2(i+1) + 1
\end{align*}
as required.
