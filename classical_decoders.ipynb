# Decoding messages

When a codeword is sent through a physical channel, some bits may be corrupted. As a result, the received message might differ from the original codeword. Without knowing exactly where the errors occured, we cannot be certain what the original codeword was. So instead we try to find the codeword that is most likely to have been sent. To do this we use a probablistic model of the transmission channel.

## Discrete channels

<div style="border:2px solid #3b82f6; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Definition. [(Binary) Discrete channel]
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">


Let $\mathcal{X}$ and $\mathcal{Y}$ be finite sets and let $X$ and $Y$ be random variables that take values in $\mathcal{X}$ and $\mathcal{Y}$ respectively. A _discrete channel_ is a set of probability mass functions 
$$p_{Y \mid X} = \{ p_{Y \mid X=x} \colon \mathcal{Y} \to [0,1] \mid x \in \mathcal{X}\}.$$
The discrete channel is denoted $(\mathcal{X}, p_{Y \mid X}, \mathcal{Y})$. A _binary_ (discrete) channel is a discrete channel with 
$$\mathcal{X} = \{0,1\} = \mathcal{Y}.$$

For each input $x \in \mathcal{X}$ and output $y \in \mathcal{Y}$, we interpret the value 
$$p_{Y \mid X=x}(y) = p(Y=y \mid X=x)$$
as the probability of observing output $y$ given that input $x$ was sent through the channel.


</div>
</div>
</div>

In practice the same physical channel is used multiple times. Ideally, to be able to analyse the outputs, we would like each use to follow the same probability distribution, However, in general the output at time $i$ could depend on past or future inputs and outputs. Fix an integer $n \in \mathbb{N}$. For each $k \in [n]$, define random variables 
$$X^k = (X_1,X_2,\dots,X_k) \;\; \text{ and } \;\; Y^k = (Y_1,Y_2,\dots,Y_k)$$
that take values in $\mathcal{X}^k$ and $\mathcal{Y}^k$ respectively. For sequences
$$ x^n = (x_1,x_2,\dots,x_n) \in \mathcal{X}^n \;\; \text{ and } \;\; y^n = (y_1,y_2,\dots,y_n) \in \mathcal{Y}^n, $$
the chain rule for conditional probabilities gives
$$ p(Y^n = y^n \mid X^n = x^n) = \prod_{i=1}^n p(Y_i = y_i \mid X^n = x^n, Y^{i-1} = y^{i-1}). $$
Thus, in order to understand the distribution of repeated uses of the channel, we need to acertain how $Y_i$ depends on the time index, all inputs $X^n$ and all previous outputs $Y^{i-1}$.  

We introduce assumptions to simplify the dependence.

<div style="border:2px solid #3b82f6; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Definition. [Causal discrete channel]
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

A discrete channel $(\mathcal{X}, p_{Y \mid X}, \mathcal{Y})$ is _causal_ if the current output does not depend on future inputs. More precisely, for any integer $n \in \mathbb{N}$ with $i \in [n]$ and sequences
$$ x^n = (x_1,x_2,\dots,x_n) \in \mathcal{X} \;\; \text{ and } \;\; y^n = (y_1,y_2,\dots,y_n) \in \mathcal{Y}, $$
 we have that 
$$ p(Y_i = y_i \mid X^n = x^n, Y^{i-1} = y^{i-1}) = p(Y_i = y_i \mid X^i = x^i, Y^{i-1} = y^{i-1}). $$

</div>
</div>
</div>

<div style="border:2px solid #3b82f6; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Definition. [Memoryless discrete channel]
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

A discrete channel $(\mathcal{X}, p_{Y \mid X}, \mathcal{Y})$ is _memoryless_ if the current output does not depend on previous inputs or outputs. More precisely, for any integer $n \in \mathbb{N}$ with $i \in [n]$ and sequences
$$ x^n = (x_1,x_2,\dots,x_n) \in \mathcal{X} \;\; \text{ and } \;\; y^n = (y_1,y_2,\dots,y_n) \in \mathcal{Y}, $$
 we have that 
$$ p(Y_i = y_i \mid X^i = x^i, Y^{i-1} = y^{i-1}) = p(Y_i = y_i \mid X_i = x_i). $$

</div>
</div>
</div>

<div style="border:2px solid #3b82f6; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Definition. [Stationary discrete channel]
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

A discrete channel $(\mathcal{X}, p_{Y \mid X}, \mathcal{Y})$ is _stationary_ if the conditional distribution of $Y_i$ on $X_i$ does not depend on $i$. More precisely, for any integer $n \in \mathbb{N}$ with $i, j \in [n]$, input $x \in \mathcal{X}$ and output $y \in \mathcal{Y}$, we have that 
$$ p(Y_i = y \mid X_i = x) =  p(Y_j = y \mid X_j = x).$$

</div>
</div>
</div>

Therefore, for a discrete channel that is causal, memoryless and stationary the conditional distribution of $Y^n$ on $X^n$ is fully determined by the conditional distribution of $Y$ on $X$. In particular, recall by the chain rule we have
$$p(Y^n = y^n \mid X^n = x^n) = \prod_{i=1}^n p(Y_i = y_i \mid X^n = x^n, Y^{i-1} = y^{i-1}).$$
Then causality gives
$$p(Y_i = y_i \mid X^n = x^n, Y^{i-1} = y^{i-1}) = p(Y_i = y_i \mid X^i = x^i, Y^{i-1} = y^{i-1})).$$
Memorylessness gives
$$p(Y_i = y_i \mid X^i = x^i, Y^{i-1} = y^{i-1})) = p(Y_i = y_i \mid X_i = x_i).$$
Finally, stationary gives
$$p(Y_i = y_i \mid X_i = x_i) = p(Y = y_i \mid X = x_i).$$
Combining these expressions we obtain
$$p(Y^n = y^n \mid X^n = x^n) = \prod_{i=1}^n p(Y = y_i \mid X = x_i).$$


### Binary symmetric channels

Since we wish to transmit binary codewords we will focus on binary channels. From now on all channels will be discrete, binary, causal, memoryless and stationary. 

Fix a channel $(\{0,1\}, p_{Y \mid X}, \{0,1\})$. The probability mass functions $p_{Y \mid X=0}$ and $p_{Y \mid X=1}$ are uniquely defined by the four probabilites
$$ p(Y=0 \mid X=0), \;\;\; p(Y=1 \mid X=0), \;\;\; p(Y=0 \mid X=1), \;\;\; p(Y=1 \mid X=1). $$
A binary symmetric channel groups these probabilities according to whether $X=Y$ or $X \neq Y$.

<div style="border:2px solid #3b82f6; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Definition. [Binary symmetric channel]
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

A binary discrete channel $(\{0,1\}, p_{X \mid Y}, \{0,1\})$ is a _binary symmetric channel_ if (it is causal, memoryless and stationary and) there exists a value $p\in [0,1]$ such that
$$p(Y = y | X = x) = \begin{cases} 1-p &\text{ if } y=x \\ p &\text{ if } y \neq x. \end{cases}$$

</div>
</div>
</div>

Consequently, for a binary symmetric channel the probability of an output sequence depends only on how many coordinates differ from the input sequence. Identifying sequences $x=(x_i)_{i=1}^n$ and $y=(y_i)_{i=1}^n$ with vectors in $\{0,1\}^n$ we can define the Hamming distance $d(x,y)$ as follows:
$$d(x,y) = \sum_{i=1}^n \mathbb{1}_{x_i \neq y_i}.$$
The sequences differ in exactly $d(x,y)$ coordinates, so we have
$$p(Y^n = y | X^n = x) = \prod_{i=1}^n p(Y=y_i \mid X = x_i) = p^d(1-p)^{n-d}.$$

## Maximum likelihood decoding

When a codeword is sent through a physical channel, some bits may be corrupted. As a result, the observed output might differ from the codeword that was sent. Decoding aims to recover the original codeword despite these errors. Without knowing the exact errors that occurred, we can never be certain that decoding will find the intended codeword. Instead we find the codeword that is most likely to have been sent with respect to the channel probabilities. 

Let $\mathcal{C}$ be a code of length $n \in \mathbb{Z}_{>0}$ and suppose the we observe an output $v \in \{0,1\}^n$. Then we wish to find a unique codeword $\hat{c} \in \mathcal{C}$ such that 
\begin{equation*}
    p(\text{ Input} = \hat{c} \mid \text{ Output} = v) = \mathrm{max}_{c \in \mathcal{C}} \{p(\text{ Input} = c \mid \text{ Output} = v)\}.
\end{equation*}

For each bit position $i \in [n]$, define two random variables:
1. Let $C_i$ be the random variable whose value is the $i^{th}$ bit of the sent codeword. Let $C$ be the random variable $C = (C_1,C_2,\dots,C_n)$, which represents the full codeword.
2. Let $V_i$ be the random variable whose value is the $i^{th}$ bit of the observed output. Let $V$ be the random variable $V = (V_1,V_2,\dots,V_n)$ which represents the full observed output.

By abuse of notation we will often drop the notation for the random variables $C$ and $V$ and instead of writing $p(C=c \mid V=v)$, for example, write $p(c \mid v)$.


The application of Bayes Theorem to $p(c \mid v)$ results in
$$p(c \mid v) = \frac{p(v \mid c)\cdot p(c)}{p(v)}.$$
Notice that $p(v)$ does not depend on the specific codeword $c$, and so, to maximise $p(c \mid v)$ over $c \in \mathcal{C}$ it suffices to maximise $$p(v \mid c)\cdot p(c)$$ over $\mathcal{C}$. Moreover, we make the assumption that each codeword is equally likely to be sent. That is, for any codeword $c \in \mathcal{C}$ we have that $$p(c) = \frac{1}{|\mathcal{C}|}.$$ Thus, maximising $p(v \mid c)\cdot p(c)$ over $c \in \mathcal{C}$ is equivalent to maximising $p(v \mid c)$ over $c \in \mathcal{C}$.


<div style="border:2px solid #7c3aed; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Example.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

Consider the repetition code of length $3$. Suppose that we send a codeword across a binary symmetric channel that has a probability $p=\frac{1}{4}$ of a bit flip and we observe the output $v=100$. We know an error has occured since $v$ is not a codeword. Therefore, we must attempt to decode $v$.

To decode using the maximum likelihood decoder we must compute $p(v \mid c)$ for each codeword $c \in \mathcal{C}$. That is, for $c=000$ and $c=111$.

Suppose the sent codeword was $c = 000$. Then observing the output $v=100$ means the an error occured in the first bit and not the second two. In particular, we have
\begin{align*}
    p(v=100 \mid c=000) &= p(v_1=1 \mid c_1=0) \cdot p(v_2=0 \mid c_2=0) \cdot p(v_3=0 \mid c_3=0) \\ &= p(v_1 \neq c_1) \cdot p(v_2 = c_2) \cdot p(v_3 = c_3) \\ &= \frac{1}{4} \cdot \frac{3}{4} \cdot \frac{3}{4} = \frac{9}{64}. 
\end{align*}
Now suppose instead that the sent codeword was $c = 111$. Then observing the output $v=100$ means an error occured in the second two bits and not the first. Thus, we have
\begin{equation*}
    p(v=100 \mid c=111) = \frac{3}{4} \cdot \frac{1}{4} \cdot \frac{1}{4} = \frac{3}{64}. 
\end{equation*}
Therefore, as $p(v=100 \mid c=000)$ is larger than $p(v=100 \mid c=111)$ it is most likely that the output $v=100$ came from the codeword $c=000$.

By repeating this process, we can compute most likely transmitted codeword for each possible output $v \in \{0,1\}^3$.

<center>

| Output $v$   | $p(v \mid c=000)$ | $p(v \mid c=111)$  | Most likely codeword $c$ |
| ---                    | ---                    | ---                    | ---             |
| 000                    | $\frac{27}{64}$        | $\frac{1}{64}$         | 000 <!-- with probability $\frac{27}{28}$ --> |
| 100, 010, 001          | $\frac{9}{64}$         | $\frac{3}{64}$         | 000 <!-- with probability $\frac{3}{4}$ --> |
| 110, 101, 011          | $\frac{3}{64}$         | $\frac{9}{64}$         | 111 <!-- with probability $\frac{3}{4}$ --> |
| 111                    | $\frac{1}{64}$         | $\frac{27}{64}$        | 111 <!-- with probability $\frac{27}{28}$ --> |

</center>
</div>
</div>
</div>

```python
import random
import time

from fractions import Fraction

from linearcodes import Codeword, LinearCode, HammingCode

def choose_random_codeword(code: LinearCode):
    k = code.rank
    B = code.basis()

    coeffs = [random.randint(0,1) for _ in range(k)]
    total = [0] * code.length

    for coeff, b_vec in zip(coeffs, B):
        if coeff:
            total = [(v ^ b) for v, b in zip(total,b_vec)]

    return Codeword(tuple(total))

def send_codeword(codeword: Codeword, channel_prob: Fraction):
    output = []
    for bit in codeword.bits:
        if random.random() < channel_prob:
            flip_bit = (bit+1)%2
            output.append(flip_bit)
        else:
            output.append(bit)
    return output

def likelihood(codeword: list[int], output: list[int], channel_prob: Fraction) -> Fraction:
    p = channel_prob
    one_minus_p = Fraction(p.denominator - p.numerator, p.denominator)

    prob = Fraction(1)

    for c, o in zip(codeword, output):
        if c == o:
            prob *= one_minus_p
        else:
            prob *= p

    return prob


def maximum_likelihood_decoder(codewords: list[list[int]], output: list[int], channel_prob: Fraction) -> dict:
    """
    Returns the dictionary which maps a codeword c to the probability p(output | input = c).
    Args:
        - channel_probabilities (list[float]):
            entry with index i is the probability bit i will flip.
        - code (lc.LinearCode):
            code used for encoding.
        - message (list[int]):
            received message.
    """

    prob_to_codewords = {}
    for c in codewords:  
        prob = likelihood(c, output, channel_prob)    
        
        if prob not in prob_to_codewords:
            prob_to_codewords[prob] = []
        
        prob_to_codewords[prob].append(c)

    return dict(sorted(prob_to_codewords.items(), reverse=True))

code = HammingCode(4)
codewords = list(code.codewords)
channel_prob = Fraction(1,3) # = 0.1

num_trials = 1
for trial in range(num_trials):
    print(f"=== Trial {trial + 1} ===")

    codeword = choose_random_codeword(code)
    output = send_codeword(codeword, channel_prob)

    start_time = time.time()
    prob_to_codewords = maximum_likelihood_decoder(codewords, output, channel_prob)
    decode_time = time.time() - start_time

    largest_p = next(iter(prob_to_codewords))
    most_likely = prob_to_codewords[largest_p]

    print(f"\nOriginal codeword   : {codeword.bits}")
    print(f"Output              : {tuple(output)}")

    print("\nMost likely codeword(s)")
    print(f"Probability p(v|c)  : {largest_p} = {float(largest_p):.3f}")
    print(f"Candidates c        : {most_likely}")

    unique = len(most_likely) == 1
    print(f"\nUnique decoding?    : {unique}")

    if unique:
        decoded = most_likely[0]
        correct = (codeword.bits == decoded)

        print(f"\nDecoded codeword    : {decoded}")
        print(f"Correct?            : {correct}")
        print(f"Decode time         : {decode_time * 1e3:.3f} × 10^{-3}")

    print("\nProbability p vs num of codewords c with p = p(v|c)")
    print("-" * 60)

    for p, cws in prob_to_codewords.items():
        print(f"P = {float(p) :.6f}  |  count = {len(cws)}")
```


### Hamming distance

Since our transmission is through a binary symmetric channel, we can already compute $p(v \mid c)$. Let $p$ be the probability a bit flips and let $d(v,c)$ denote the Hamming distance. Then we have
$$p(v \mid c) = p^{d(v,c)}(1-p)^{n-d(v,c)}.$$

Rearranging this expression we obtain
$$p(v \mid c) = \left( \frac{p}{1-p}\right)^{d(v,c)}  \cdot (1-p)^{n}.$$
Notice that $(1-p)^{n}$ depends neither on the codeword $c$ nor the observed output $v$. Thus, maximising $p(v \mid c)$ over $c \in \mathcal{C}$ is equivalent to maximising
$$\left( \frac{p}{1-p}\right)^{d(v,c)}$$
over $c \in \mathcal{C}$.

By convention we assume that $p < \frac{1}{2}$. If it is not, then we can assume the received message is more likely to be wrong and flip all the bits before working with them .... (Fix this phrasing!). Consequently, the value $\frac{p}{1-p}$ is less than $1$ and so to maximise $$\left( \frac{p}{1-p}\right)^{d(v,c)}$$ we must minimise $d(v,c)$. In particular, the codeword $\hat{c}$ that maximises $p(c \mid v)$ is precisely the codeword that minimises $d(v,c)$. That is, the codeword that is closest to $v$.

<div style="border:2px solid #7c3aed; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Example.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

We recompute the maximum likelihood decoding for the repetition code using the Hamming distance. Recall we use the repetition code of length $3$ and observe the output $v=100$. There are two codewords in the repetition code. We can compute their Hamming distance to $v$ as follows:
$$
d(v=100,\, c=000) = 1 \;\; \text{and} \;\; d(v=100,\, c=111) = 2.
$$
Therefore, with respect to the Hamming distance, $c = 000$ is the closest codeword to $v$.

We can repeat this computation for all possible outputs $v \in \{0,1\}^3$.

<center>

| Output $v$   | $d(v,\, c=000)$ | $d(v,\, c=111)$  | Most likely codeword $c$ |
| ---                    | ---       | ---       | ---             |
| 000                    | 0         | 3         | 000 <!-- with probability $\frac{27}{28}$ --> |
| 100, 010, 001          | 1         | 2         | 000 <!-- with probability $\frac{3}{4}$ --> |
| 110, 101, 011          | 2         | 1         | 111 <!-- with probability $\frac{3}{4}$ --> |
| 111                    | 3         | 0         | 111 <!-- with probability $\frac{27}{28}$ --> |

</center>

**Example.** By using the Hamming distance we save a massive amount of computational time.

```python
from linearcodes import HammingCode

import time

def hamming_distance(c: list[int], y: list[int]) -> int:
    if len(c) != len(y):
        raise ValueError("Both x and y must be the same length.")
    return sum([1 for i in range(len(c)) if c[i] != y[i]])


def minimise_hamming_distance(codewords: list[list[int]], output: list[int]) -> dict:
    """
    Returns the codeword that was most likely to be sent given message is received using Hamming distance.
    Args:
        - channel_probabilities (list[float]):
            entry with index i is the probability bit i will flip.
        - code (lc.LinearCode):
            code used for encoding.
        - message (list[int]):
            received message.
    """
    hamming_to_codewords = {}
    for c in codewords:  
        d = hamming_distance(c, output)    
        
        if d not in hamming_to_codewords:
            hamming_to_codewords[d] = []
        
        hamming_to_codewords[d].append(c)

    return dict(sorted(hamming_to_codewords.items()))

code = HammingCode(6)
codewords = list(code.codewords)
channel_prob = Fraction(1,10) # = 0.1

num_trials = 1
for trial in range(num_trials):
    print(f"=== Trial {trial + 1} ===")

    codeword = choose_random_codeword(code)
    output = send_codeword(codeword, channel_prob)

    start_time = time.time()
    hamming_to_codewords = minimise_hamming_distance(codewords, output)
    decode_time = time.time() - start_time

    smallest_d = next(iter(hamming_to_codewords))
    most_likely = hamming_to_codewords[smallest_d]

    print(f"\nOriginal codeword   : {codeword.bits}")
    print(f"Output              : {tuple(output)}")

    print("\nMost likely codeword(s)")
    print(f"Distance d(v,c)     : {smallest_d}")
    print(f"Candidates c        : {most_likely}")

    unique = len(most_likely) == 1
    print(f"\nUnique decoding?    : {unique}")

    if unique:
        decoded = most_likely[0]
        correct = (codeword.bits == decoded)

        print(f"\nDecoded codeword    : {decoded}")
        print(f"Correct?            : {correct}")
        print(f"Decode time         : {decode_time * 1e3:.3f} × 10^{-3}")

    print("\nDistance d vs num of codewords c with d = d(v,c)")
    print("-" * 60)

    for d, cws in hamming_to_codewords.items():
        print(f"Distance = {d}  |  count = {len(cws)}")
```

The Hamming distance not only provides a more efficient way to compute the most likely codeword, it also determines how many errors a code can detect and correct in general.

<div style="border:2px solid #22c55e; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Lemma.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">



Let $\mathcal{C}$ be a linear code with distance $d$. Then $\mathcal{C}$ can detect $d-1$ errors and correct $\lfloor \frac{d-1}{2} \rfloor$ errors.


</div>
</div>
</div>

<div style="border:2px solid #86efac; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Proof.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

Let $v$ be the observed output from a codeword $c$. Then $d(v,c)$ is the number of transmission errors. If no errors occured, then the statements hold vacuosly, so suppose that $v \neq c$.

The only way we cannot detect any errors in the observed output $v$ is if $v$ is a codeword itself. Since $v \neq c$, it follows that $v$ is another distinct codeword and we have that $d(v,c) \geq d$ and at least $d$ errors occured in $v$. Consequently, if $d(c,v) < d$, then $v$ cannot be a codeword and we can detect any errors in $v$.

To be able to correct the errors in $v$, we need $v$ to have a unique nearest codeword. Suppose there exists a codeword $c' \in \mathcal{C}$ such that $c' \neq c$ and
$$d(v,c) = d(v,c').$$
By the triangle inequality, we have
$$ d(c,c') \leq d(c,v) + d(v,c'). $$
Since $d \leq d(c,c')$ we rearrange this equation to obtain
$$ d \leq d(c,c') \leq d(c,v) + d(v,c') = 2d(v,c).$$
Therefore, we have that $d(v,c) \geq \frac{d}{2}$. Consequently, if 
$$d(v,c) < \frac{d}{2},$$
then $c$ is the unique closest codeword to $v$. That is, we can correct $\lfloor \frac{d-1}{2} \rfloor$ errors.

</div>
</div>
</div>

No code will ever be able to correct all errors, it is about reducing the probability of error.

<div style="border:2px solid #22c55e; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Theorem. [Hamming bound]
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

Let $\mathcal{C}$ be a linear code of length $n$ and rank $k$. Fix an integer $t \in [n]$. If the code $\mathcal{C}$ can correct $t$ errors, then
$$ k \leq n - \mathrm{log}_2 \left[\sum_{r=0}^t \binom{n}{r} \right]. $$


</div>
</div>
</div>

<div style="border:2px solid #86efac; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Proof.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">


For each codeword $c \in \mathcal{C}$ and integer $r \in [n]$, define the Hamming ball
$$ B_r(c) = \{v \in \{0,1\}^n \mid d(c,v) \leq r \}.$$
A code can correct $t$ errors only if every observed output $v \in \{0,1\}^n$ has a unique codeword $c \in \mathcal{C}$ such that $d(c,v) \leq t$. That is, each output $v$ lies in a unique ball $B_t(c)$. Equivalently, the Hamming balls
$$ \{ B_t(c) \mid c \in \mathcal{C} \} $$
must be disjoint. Since the ambient space contains $2^n$ vectors, we obtain a bound
$$ \sum_{c \in \mathcal{C}} | B_t(c) | \leq 2^n. $$

Now we compute the size of each Hamming ball. Fix a codeword $c \in \mathcal{C}$ and a radius $r \in [n]$. The Hamming distance between two vectors is $r$ if and only if they differ in exactly $r$ coordinates. For a vector of length $n$, there are $\binom{n}{r}$ ways to uniquely choose these $r$ coordinates. Thus, the size of the ball $B_t(c)$ is 
$$ | B_t(c) | = \sum_{r=0}^t \binom{n}{r}. $$
In particular, the size does not depend on the codeword $c$.

Therefore, we have 
$$ |\mathcal{C}| \cdot{} \sum_{r=0}^t \binom{n}{r} = \sum_{c \in \mathcal{C}} | B_t(c) | \leq 2^n. $$
For a code of rank $k$, there are $2^k$ codewords, so we obtain 
$$ 2^k \cdot{} \sum_{r=0}^t \binom{n}{r} = |\mathcal{C}| \cdot{} \sum_{r=0}^t \binom{n}{r} \leq 2^n. $$
Taking logs of both sides we obtain 
$$ k + \log_2\left[\sum_{r=0}^t \binom{n}{r} \right] \leq n. $$


</div>
</div>
</div>

## Message Passing

Maximum likelihood decoding identifies the most likely codeword for a received message by comparing it against every possible codeword. While this is theoretically optimal, it becomes infeasible for practical codes because the number of codewords grows exponentially with the code rank. Message passing is a scalable alternative where instead of trying to identify which codeword is the most likely to be sent, we instead identify for each bit position which bit was most likely to have been sent. In general, these two probabilities do not coincide. That is



It is a graph-based algorithm where each node has local information and can communicate only with its immediate neighbours. Nodes iteratively update their knowledge based on messages received, and through repeated exchanges, information spreads across the network. Over time, global properties can be inferred even though no node initially has complete knowledge.

Message passing is a analogy of social networks. People who spend a lot of time together tend to share opinions and information. Each individual updates their understanding based on what they hear from friends. Then when they passes this updated information along to other friends. As this process iterates, ideas and knowledge propagate through the network, allowing people to build a collective understanding even without direct access to everyone’s thoughts.

We can use this type of algorithm on the Tanner graph of an error correcting code. Bit nodes and parity check nodes will exchange messages along edges of the graph, updating their beliefs about errors iteratively. First we do a concrete example of message passing.

### Think of a title

Suppose we wish to count the number of vertices in a tree. Each node can only see itself and its neighbours, so it only has local information. However, we can use message passing to combine this information and create a global picture.

Let $V$ be the set of vertices and $E$ be the set of edges of a graph $G=(V, E)$. For a vertex $u \in V$ and let $\mathcal{N}(u)$ denote the neighbourhood of the vertex $u$ in $G$. For each vertex $u \in V$, we define a sequence integers $$N_0(u), \, N_1(u), \, N_2(u), \, N_3(u) \, \dots$$ by the following algorithm:

**Initialisation:** For each vertex $u \in V$, define $N_0(u) = 1$. For each pair of neighbouring vertices $(u,v)$ define $m_1(u,v) = 1$.

**Iterative step for $i>0$:**
1. **Update vertex values:** Update the information for each vertex using the following formula
$$N_i(u) = N_{i-1}(u) + \sum_{v \in \mathcal{N}(u)} m_{i}(v,u).$$
2. **Update messages:** For each pair of neighbouring vertices $(u,v)$ define a message $m_{i+1}(u,v)$ by
    $$ m_{i+1}(u,v) = \sum_{\substack{w \in \mathcal{N}(u) \\ w \neq v}} m_{i}(w,u). $$

    The message $m_{i+1}(u,v)$ will be the information "sent" from vertex $u$ to vertex $v$. This is why we do not include $v$ in the summation index for $m_i(u,v)$. If we did we would be sending information back to $v$ that it already knows and sent to us on the previous iteration.
3. **Check stop condition:** If all the messages are $0$, stop the algorithm.
 
**Claim:** Let $G = (V,E)$ be a finite tree. Then the algorithm terminates after finitely many steps. Moreover, if the algorithm terminates on iteration $i>0$, then for each vertex $u \in V$, the value $N_i(u)$ is the number of vertices in $T$.

Before, proving this statement, let us consider a concrete example.

Consider an $A_3$ graph, that is a straight line with three vertices
$$ u_1 \text{---} u_2 \text{---} u_3. $$
Each of $N_0(u_i)$ is initialised at $1$ and each message is initialised at $1$. Now let us apply the algorithm.

**Iteration 1**
1. **Update vertex values:** We have that
    $$ N_1(u_1) = N_0(u_1) + \sum_{v \in \mathcal{N}(u_1)} m_1(v,u_1) = N_0(u_1) + m_1(u_2, u_1) = 1 + 1 = 2.$$
    Similarly, we compute that $N_1(u_2) = 3$ and $N_1(u_3)=1$.
2. **Update messages:** Since $u_1$ and $u_3$ have exactly one neighbour, namely $u_2$, we have that
    $$m_2(u_1,u_2) = \sum_{\substack{w \in \mathcal{N}(u_1) \\ w \neq u_2}} m_{1}(w,u_1) = 0$$
    and similarly $m_2(u_3,u_2) = 0$. However, we have two non-zero messages given by
    $$m_2(u_2,u_1) = \sum_{\substack{w \in \mathcal{N}(u_2) \\ w \neq u_1}} m_{1}(w,u_2) = m_{1}(u_3,u_2)=1$$
    and $m_2(u_2,u_3) = m_1(u_1,u_2) = 1$.
3. **Check stop condition:** Two of the messages are non-zero.

**Iteration 2**
1. **Update vertex values:** Since $u_2$ only receives zero messages, it follows that $N_2(u_2) = N_1(u_2) = 3$. For $u_1$ we have that
    $$ N_2(u_1) = N_1(u_1) + \sum_{v \in \mathcal{N}(u_1)} m_2(v,u_1) = N_1(u_1) + m_2(u_2, u_1) = 2 + 1 = 3.$$
    Similarly, we compute that $N_2(u_3) = 3$.
2. **Define messages:** All the messages will be zero in this iteration. Again since $u_1$ and $u_3$ have exactly one neighbour, namely $u_2$, we have that $m_3(u_1,u_2) = 0$ and $m_3(u_3,u_2)=0$. Moreover, we have that
    $$m_3(u_2,u_1) = \sum_{\substack{w \in \mathcal{N}(u_2) \\ w \neq u_1}} m_{1}(w,u_2) = m_2(u_3,u_2)=0$$
    and $m_3(u_2,u_3) = m_2(u_1,u_2) = 0$.
3. **Check stop condition:** All the messages are $0$. Stop the algorithm.

The algorithm terminated on iteration $2$, so for each vertex we obtained a sequence of $3$ values, as follows
$$ u_1: \, N_0(u_1) = 1, \, N_1(u_1) = 2, \, N_2(u_1) = 3, $$
$$ u_2: \, N_0(u_2) = 1, \, N_1(u_2) = 3, \, N_2(u_2) = 3, $$
$$ u_3: \, N_0(u_3) = 1, \, N_1(u_3) = 2, \, N_2(u_3) = 3. $$
The final value of each sequence is $3$; the number of vertices in $A_3$.

We prove this Theorem as a series of Lemmas.

<div style="border:2px solid #22c55e; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Lemma.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">


Let $G = (V,E)$ be a tree. Fix vertices $u, v \in V$ and an integer $i>0$. Then the message $m_i(u,v)$ is the number of vertices whose minimum distance from $u$ is $i-1$ via a path that does not go through $v$.

</div>
</div>
</div>

<div style="border:2px solid #86efac; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Proof.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">



We prove this by induction on the number of iterations. By definition, $m_1(u,v) = 1$. This correctly counts the vertex $u$ itself, which is at distance $0 = i-1$ from $u$ and whose minimal path does not go through $v$. Now assume the claim holds for iteration $i$ and consider iteration $i+1$. By the message passing algorithm, we have
$$
m_{i+1}(u,v) = \sum_{w \in \mathcal{N}(u), w \neq v} m_i(w,u).
$$
By the induction hypothesis, each $m_i(w,u)$ counts all vertices whose minimal path from $w$ does not go through $u$ and has length $i-1$. Concatenating each of these paths with the edge $w \to u$ gives paths of length $i$ from $u$ that still do not pass through $v$. Moreover, since $G$ is a tree, there is a unique path between any two vertices. Therefore, summing over all neighbours $w \neq v$ of $u$ correctly counts all vertices whose minimal path from $u$ has length $i$ and does not pass through $v$.


</div>
</div>
</div>

<div style="border:2px solid #22c55e; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Lemma.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">

Let $G = (V,E)$ be a tree. Fix a vertex $u \in V$ and an integer $i>0$. Then $N_i(u)$ is the number of vertices that are at most $i$ steps away from $u$.

</div>
</div>
</div>

<div style="border:2px solid #86efac; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Proof.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">


We prove this by induction on the number of iterations. By definition, $N_0(u) = 1$, which correctly counts only the vertex $u$ itself, as it is the only vertex at distance $0$ from $u$. Assume the statement holds for iteration $i$ and consider iteration $i+1$. By the message passing algorithm,
$$
N_{i+1}(u) = N_i(u) + \sum_{v \in \mathcal{N}(u)} m_i(v,u).
$$
By the induction hypothesis, $N_i(u)$ counts all vertices at distance at most $i$ from $u$. Moreover, for each neighbour $v \in \mathcal{N}(u)$, the message $m_i(v,u)$ counts all vertices whose minimal path from $v$ does not go through $u$ and has length $i$. Therefore, concatenating these paths with the edge $v \to u$ gives a new path to vertices at distance $i+1$ from $u$. Since $G$ is a tree, there is a unique path between any two vertices. Therefore, summing over all neighbours $v$ correctly counts all vertices that are exactly $i+1$ steps away from $u$. Adding $N_i(u)$ then counts all vertices at distance at most $i$ from $u$.


</div>
</div>
</div>

<div style="border:2px solid #22c55e; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Lemma.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">



Let $G = (V,E)$ be a finite tree. Then the algorithm terminates after finitely many steps. Moreover, if the algorithm terminates on iteration $i>0$, then for each vertex $u \in V$, the value $N_i(u)$ is the number of vertices in $T$.


</div>
</div>
</div>

<div style="border:2px solid #86efac; border-radius:8px; max-width:95%; margin:auto; overflow:hidden;">

<div style="padding:8px 12px;">
Proof.
</div>

<div style="background: display:flex; justify-content:center;">

<div style="max-width:95%; line-height:1.6; padding:12px;">


Since $G$ is a finite tree, there exists an integer $d>0$ such that the distance between any two vertices in $G$ is at most $d$. Recall, for any vertices $u$ and $v$, the message $m_{d+1}(u,v)$ counts the number of vertices whose minimal path from $u$ is of length $d$ and does not go through $v$. If such a vertex exists, then concatenating its minimal path to $u$ with the edge $u \text{---} v$ would create a path of length $d+1$; a contradiction. Therefore, for all vertices $u$ and $v$, we have $m_{d+1}(u,v)=0$ and the algorithm terminates at iteration $d$. Moreover, because $N_d(u)$ is the number of vertices at most $d$ steps away from $u$, we have that $N_d(u) = |V|$.

We can encode message passing to work on large graphs.

```python
import networkx as nx
import random

class MessagePassing:
    """
    Implements a message passing algorithm on a NetworkX tree.
    Algorithm sends integer messages, adding them at each iteration.
    
    Args:
        - num_vertices [int]: Number of vertices in the tree.
        - initial_vertex_values [list[int]]: Optional list of initial values for each vertex.
            If not given, then defaults to 1 for all vertices.
    """
    def __init__(self,
                 graph: nx.Graph,
                 initial_vertex_values: list[int] | None = None):
        self.graph = graph
        self.num_vertices = len(graph.nodes)  

        if initial_vertex_values is None:
            initial_vertex_values = [1]*self.num_vertices
        if len(initial_vertex_values) != self.num_vertices:
            raise ValueError("Length of initial_vertex_values must match num_vertices.")
        
        self.neighbours = [list(self.graph.neighbors(u)) for u in self.graph.nodes]
        self.initial_vertex_values = initial_vertex_values
        
        self.directed_edges = [(u, v) for u in self.graph.nodes for v in self.graph.neighbors(u)]
        
    @property
    def initial_messages(self) -> dict[tuple[int,int],int]:
        """
        Creates list of initial messages.
        The message from a vertex u to a neighbour v is the inital value of u.
        """
        return {edge : self.initial_vertex_values[edge[0]] for edge in self.directed_edges}

    def run(self, max_interations: int = 100, verbose: bool = False) -> tuple[int,list[int]]:
        """
        Run message passing until convergence or max_iterations reached.
        
        Args:
            - max_iterations [int]: Maximum number of iterations of the algorithm.
            - verbose [bool]: If True, prints messages and vertex values at each iteration.
        
        Returns:
            list[int]: The final vertex values after message passing.
        """
        vertex_values = self.initial_vertex_values
        messages = self.initial_messages.copy()
        for iteration in range(max_interations):
            vertex_values = self._update_vertex_values(vertex_values, messages)
            
            new_messages = self._update_messages(messages)
            if all(value == 0 for value in new_messages.values()):
                    if verbose:
                        print(f"Converged at iteration {iteration+1}.")
                    return (iteration, vertex_values)
            messages = new_messages

            if verbose:
                print(f"\nIteration {iteration+1}")
                print(f"Messages: {messages}")
                print(f"Vertex values: {vertex_values}")

        return (iteration,vertex_values)

    def _update_messages(self,
                        messages: dict[tuple[int,int], int]) -> dict[tuple[int,int], int]:
        """
        Compute new messages based on incoming messages.
        """
        new_messages = {edge : 0 for edge in self.directed_edges}
        for u, v in self.directed_edges:
            new_message = 0
            for w in self.neighbours[u]:
                if w != v:
                    try:
                        new_message += messages[(w,u)]
                    except:
                        pass
            new_messages[(u,v)] = new_message
        return new_messages
    
    def _update_vertex_values(self,
                              vertex_values: list[int],
                              messages: dict[tuple[int,int], int]) -> list[int]:
        """
        Update vertex values based on the messages received from neighbours.
        """
        new_vertex_values = vertex_values.copy()
        for u, v in self.directed_edges:
            new_vertex_values[v] += messages[(u,v)]
        return new_vertex_values

def find_number_vertices_in_random_tree(max_num_vertices: int = 50, num_iterations: int = 100):
    num_vertices = random.choice(range(max_num_vertices))
    graph = nx.generators.random_labeled_tree(num_vertices)
    mp = MessagePassing(graph)
    final_info = mp.run(num_iterations)
    return (final_info[0], final_info[1][0])        
    
for _ in range(10):
    iteration, num_vertices = find_number_vertices_in_random_tree(1000, 1000)
    print(f"After {iteration} iterations computed the tree has {num_vertices} vertices.")
```

If the graph is not a tree, then the algorithm is not guaranteed to converge in finitely many steps.

Consider the complete graph on $3$ vertices $V = \{u_1, u_2, u_3\}$. Then we claim that for each pair of vertices $u_k, u_\ell \in V$ and each iteration $i>0$, we have that 
$$m_i(u_k,u_\ell)=1 \; \text{ and } \; N_i(u_k) = 2i + 1.$$

*Proof:* We shall prove the statement by induction on the number of iterations $i$. Each of the vertices is initialised at $1$ which coincides with $2(0) + 1 = 1$ and the first messages are also initialised at $1$ as required. Now suppose the statement holds for iteration $i$ and consider iteration $i+1$. Since the graph is cyclically symmetric it suffices to prove the statement for just one of the messages and one of the vertex values. Let's consider $m_{i+1}(u_1,u_2)$ and $N_{i+1}(u_1)$. The updated message $m_{i+1}(u_1,u_2)$ is given by the formula
\begin{align*}
m_{i+1}(u_1, u_2) &= \sum_{w \in \mathcal{N}(u_1) \setminus \{u_2\}} m_i(w, u_1) \\ &= m_i(u_3, u_1).
\end{align*}
Moreover, by the inductive hypothesis, we have that $m_i(u_3,u_1) = 1$, and so $m_{i+1}(u_1,u_2) = 1$ as required. Now consider the value of the vertex $u_1$. We have that 
\begin{align*}
N_{i+1}(u_1) &= N_i(u_1) + \sum_{v \in \mathcal{N}(u_1)} m_{i+1}(v, u_1) \\ &= N_i(u_1) + m_{i+1}(u_2, u_1) + m_{i+1}(u_3, u_1).
\end{align*}
By the inductive hypothesis, we have that $N_i(u) = 2i+1$ and $m_i(u_2, u_1) = m_i(u_3, u_1) = 1$. Thus, we obtain
\begin{align*}
N_{i+1}(u_1) &= N_i(u_1) + m_i(u_2, u_1) + m_i(u_3, u_1) \\ &= (2i + 1) + 1 + 1 = 2(i+1) + 1
\end{align*}
as required.
